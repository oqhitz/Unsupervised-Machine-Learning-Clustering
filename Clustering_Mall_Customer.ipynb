## Clustering Analysis

```python
df_cluster= data.iloc[:, [2, 3, 4]]
df_cluster
```

## Kmeans Algorithm

**The Elbow Method to find the No. of Optimal Clusters**

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters = i, init = 'k-means++', random_state = 123)
    km.fit(df_cluster)
    wcss.append(km.inertia_)
    
plt.figure(figsize=(15,5))
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method', fontsize = 20)
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show()
```

**Visualizaing the Clusters**

```python
km = KMeans(n_clusters = 5, init = 'k-means++', random_state = 123)
cluster = km.fit_predict(df_cluster)
```

```python
cluster
```

```python
df_cluster['cluster']=cluster
```

```python
hasil1=df_cluster.groupby(['cluster']).agg({"Age":"mean","Annual Income (k$)":"mean","Spending Score (1-100)":"mean"}).reset_index()
hasil1
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x=df_cluster['cluster'],y=df_cluster['Annual Income (k$)'])
plt.title('K Means Clustering', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x=df_cluster['cluster'],y=df_cluster['Spending Score (1-100)'])
plt.title('K Means Clustering', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x=df_cluster['cluster'],y=df_cluster['Age'])
plt.title('K Means Clustering', fontsize = 20)
plt.show()
```

## Hierarchial Clustering

**Using Dendrograms to find the no. of Optimal Clusters**

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(15,10))
dendrogram = sch.dendrogram(sch.linkage(df_cluster.drop(labels=['cluster'],axis=1), method = 'ward'))
plt.title('Dendrogam', fontsize = 20)
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

**Visualizing the Clusters of Hierarchial Clustering**

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
cluster_hc = hc.fit_predict(df_cluster.drop(labels=['cluster'],axis=1))
```

```python
cluster_hc
```

```python
df_cluster['cluster_hc']=cluster_hc
```

```python
df_cluster
```

```python
hasil2=df_cluster.groupby(['cluster_hc']).agg({"Age":"mean","Annual Income (k$)":"mean","Spending Score (1-100)":"mean"}).reset_index()
hasil2
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x=df_cluster['cluster_hc'],y=df_cluster['Annual Income (k$)'])
plt.title('Hierarchical Clustering', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x=df_cluster['cluster_hc'],y=df_cluster['Spending Score (1-100)'])
plt.title('Hierarchical Clustering', fontsize = 20)
plt.show()
```
